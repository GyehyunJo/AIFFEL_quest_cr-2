```python
import random

class Account:    # 클래스 Account 선언
    account_count = 0

    def __init__(self, name, balance):
        self.deposit_count = 0
        self.withdraw_count = 0
        self.deposit_log = []
        self.withdraw_log = []

        self.name = name    # 예금주 설정
        self.balance = balance    # 잔액 설정
        self.bank = "SC은행"    # 은행이름 설정

        num1 = random.randint(0, 999)    # 계좌번호 3자리-2자리-6자리로 랜덤 생성
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)
        self.account_number = num1 + '-' + num2 + '-' + num3

        Account.account_count += 1

    def get_account_num(cls):    # 클래스 변수 출력
        print(cls.account_count)

    def deposit(self, amount):    # 입금 메서드
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount)

            self.deposit_count += 1    # 입금 횟수 증가
            print(f"{self.deposit_count}회 : 입금    금액 : {amount:,}    잔액 : {self.balance:,}")    # 입금 로그 출력
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                print(f"{self.deposit_count}회 : 이자    금액 : {amount:,}    잔액 : {self.balance:,}")    # 이자 지급 메시지
        else:
            print("입금은 최소 1원 이상만 가능합니다.")    # 최소 입금 금액 메시지

    def withdraw(self, amount):    # 출금 메서드
        if self.balance >= amount:
            self.balance -= amount
            self.withdraw_log.append(amount)
            self.withdraw_count += 1    # 출금 횟수 증가
            print(f"{self.withdraw_count}회 : 출금    금액 : {amount:,}    잔액 : {self.balance:,}")    # 출금 로그 출력
        else:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")    # 잔고 부족 메시지

    def display_info(self):    # 정보 출력 메서드
        print(f"은행이름 : , {self.bank}, 예금주 : {self.name}, 계좌번호 : {self.account_number}, 잔고 : {self.balance:,}")

    def deposit_history(self):
        print("입금 기록 : ")
        for amount in self.deposit_log:
            print(f"\t{amount:,}원")    # 탭 추가

    def withdraw_history(self):
        print("출금 기록 : ")
        for amount in self.withdraw_log:
            print(f"\t{amount:,}원")    # 탭 추가

data = []
first = Account("임만순", 20000000)
second = Account("이도", 100000)
third = Account("차정은", 1000000)

data.append(first)    # 여러 객체 생성
data.append(second)
data.append(third)

first.withdraw(5000000)    # 출금
first.deposit(1000000)    # 입금
first.deposit(1000000)
first.deposit(1000000)
first.deposit(1000000)
first.deposit(1000000)
first.display_info()    # 계좌 정보 출력
first.deposit_history()  # 입금 기록 출력
first.withdraw_history() # 출금 기록 출력
print("-------------------------")
second.withdraw(100000)    # 출금
second.deposit(100000)    # 입금
second.deposit(100000)
second.deposit(100000)
second.deposit(100000)
second.deposit(100000)
second.display_info()    # 계좌 정보 출력
second.deposit_history()  # 입금 기록 출력
second.withdraw_history() # 출금 기록 출력
print("-------------------------")
third.withdraw(1000000)    # 출금
third.deposit(1000000)    # 입금
third.deposit(1000000)
third.deposit(1000000)
third.deposit(1000000)
third.deposit(1000000)
third.display_info()    # 계좌 정보 출력
third.deposit_history()  # 입금 기록 출력
third.withdraw_history() # 출금 기록 출력
print("-------------------------")
print("계좌 개수")
first.get_account_num()    # 계좌 개수
print("-------------------------")
print("[ 잔고 100만원 이상 고객 ]")

for c in data:
    if c.balance >= 1000000:    # 잔고 100만원 이상 고객 정보
        c.display_info()
```
