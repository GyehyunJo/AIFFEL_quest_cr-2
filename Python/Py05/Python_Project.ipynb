```python
# 프로젝트 1 : 사각형 넓이 구하기
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴\n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            print(self.rect())  # rect() 메서드 호출

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            print(self.par())  # par() 메서드 호출

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            print(self.trape())  # trape() 메서드 호출

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. (예시 : 가로,세로)\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, height = map(int, input('밑변과 높이를 입력하세요. (예시 : 밑변,높이)\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        base1, base2, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. (예시 : 윗변,아랫변,높이)\n >>>').split(','))
        area = ((base1 + base2) / 2) * height
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

# 객체 생성
a = Square()  # 객체 생성 후 도형의 넓이를 계산합니다.
```

```python
# 프로젝트 2 : 키오스크 주문 및 추가 주문
class Kiosk:
    def __init__(self):
        self.menu = ['아메리카노', '카페라떼', '녹차', '홍차', '모카']
        self.price = [3000, 3500, 4000, 4500, 5000]
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.price_sum = 0  # 합계 금액 초기화

    def menu_print(self):
        print("메뉴:")
        for i, item in enumerate(self.menu):
            print(f"{i + 1}. {item} - {self.price[i]}원")

    def menu_select(self):
        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료를 번호를 입력하세요 : "))  # 음료 번호 입력

            if 1 <= n <= len(self.menu):
                self.order_price.append(self.price[n - 1])  # 가격 리스트에 추가합니다.
                self.price_sum += self.price[n - 1]  # 합계 금액 업데이트

                # 음료 온도 물어보기
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_menu.append(self.temp + ' ' + self.menu[n - 1])  # 주문 리스트에 추가합니다.
                print(self.temp, self.menu[n - 1], ' : ', self.price[n - 1], '원')  # 주문 결과 출력

                # 추가 주문 또는 지불
                while True:
                    print()  # 줄 바꾸기
                    n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
                    if n > 0 and n <= len(self.menu):
                        # 추가 음료 온도
                        t = 0
                        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                            if t == 1:
                                self.temp = 'HOT'
                            elif t == 2:
                                self.temp = 'ICE'
                            else:
                                print("1과 2 중 하나를 입력하세요.\n")

                        self.order_menu.append(self.temp + ' ' + self.menu[n - 1])  # 추가 음료 주문 리스트에 추가
                        self.order_price.append(self.price[n - 1])  # 추가 음료 가격 리스트에 추가
                        self.price_sum += self.price[n - 1]  # 합계 금액 업데이트

                        print('추가 주문 음료', self.temp, self.menu[n - 1], ':', self.price[n - 1], '원\n', '합계 : ', self.price_sum, '원')
                    else:
                        if n == 0:  # 지불을 입력하면
                            print("주문이 완료되었습니다.")
                            print("주문 목록:", self.order_menu)
                            print("가격 목록:", self.order_price)
                            print("총 합계:", self.price_sum, '원')
                            return  # 함수 종료
                        else:  # 없는 번호를 입력할 때
                            print("없는 메뉴입니다. 다시 주문해 주세요.")
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

# 객체 생성 및 실행
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
```

```python
# 프로젝트 3 : 키오스크 지불 및 주문표 작성
class Payment:
    def __init__(self, amount):
        self.amount = amount

    def pay(self, method):
        print(f"{method}로 {self.amount}원을 결제합니다.")
        # 결제 처리 로직 추가 가능

class Kiosk:
    def __init__(self):
        self.menu = ['HOT americano', 'ICE latte', 'ICE mocha', 'ICE choco_latte']
        self.prices = [2000, 3000, 3000, 3000]
        self.selected_items = []  # 주문한 상품 리스트
        self.total_amount = 0  # 총 금액

    def menu_print(self):
        print("메뉴:")
        for index, item in enumerate(self.menu, start=1):
            print(f"{index}. {item} - {self.prices[index - 1]}원")

    def menu_select(self):
        while True:
            try:
                choice = int(input("주문할 메뉴의 번호를 선택하세요 (0을 입력하면 종료): "))
                if choice == 0:
                    break
                if 1 <= choice <= len(self.menu):
                    self.selected_items.append(self.menu[choice - 1])
                    self.total_amount += self.prices[choice - 1]
                    print(f"{self.menu[choice - 1]}를 선택하셨습니다.")
                else:
                    print("잘못된 선택입니다. 다시 시도하세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해 주세요.")

    def pay(self):
        if self.selected_items:
            payment_method = input("결제 수단을 선택하세요 (1: 현금, 2: 카드): ")
            if payment_method == '1':
                method = "현금"
            elif payment_method == '2':
                method = "카드"
            else:
                print("잘못된 선택입니다. 결제를 취소합니다.")
                return

            # 결제 메서드 호출
            payment = Payment(self.total_amount)
            payment.pay(method)
        else:
            print("먼저 메뉴를 선택하세요.")

    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for item in self.selected_items:
            index = self.menu.index(item)  # 선택한 상품의 인덱스 찾기
            print(f"{item} : {self.prices[index]}원")

        print('합계 금액 :', self.total_amount, '원')

        # 외곽
        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

# 실행 예시
if __name__ == "__main__":
    a = Kiosk()  # 객체 생성
    a.menu_print()  # 메뉴 출력
    a.menu_select()  # 주문
    a.pay()  # 지불
    a.table()  # 주문표 출력
```
